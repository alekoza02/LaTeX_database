```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from scipy.ndimage import maximum_filter

img = Image.open("data_raw.png")
img = np.array(img)
img = img[:,:,0] / 255

img_thresh = img > 0.9

neighborhood_dim = 50
neighborhood = [[True for j in range(neighborhood_dim)] for i in range(neighborhood_dim)]
neighborhood = np.array(neighborhood)

maximum = maximum_filter(img, footprint=neighborhood) == img
maximum_thresh = maximum.copy()
maximum_thresh &= img_thresh

plt.imshow(img)
plt.show()

plt.imshow(maximum, cmap="Grays")
plt.show()
# plt.savefig('filter_nothresh.png', dpi=300)

plt.imshow(maximum_thresh, cmap="Grays")
plt.show()
# plt.savefig('filter_sithresh.png', dpi=300)
```
